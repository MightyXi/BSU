```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("homicide-data.csv")
df.head()
```

```python
df.drop(df[df['victim_age'] == 'Unknown'].index, inplace = True)
df.drop(df[df['victim_sex'] == 'Unknown'].index, inplace = True)
df.drop(df[df['victim_race'] == 'Unknown'].index, inplace = True)

df.info()
```

```python
df['victim_age'] = pd.to_numeric(df['victim_age'])
```

```python
sns.countplot(data=df, x='victim_race')
```

```python
df['victim_sex'].value_counts()
df.victim_sex.value_counts().plot(kind='pie')
```

```python
sns.countplot(data=df, x='victim_race', palette='autumn', hue='victim_sex')
```

```python
df.victim_age.plot(kind='hist',bins=50,color='green')
```

```python
sns.displot(df, x="victim_age", bins=50, hue="victim_sex")
```

```python
sns.countplot(data=df, x='disposition')
```

```python
sns.countplot(data=df, x='state', palette='autumn', hue='victim_race')
```

```python
sns.countplot(data=df, x='state', palette='autumn', hue='disposition')
```

```python
chicago = df[ df['city'] == 'Chicago']
sns.scatterplot(data=chicago, x='lat', y='lon', hue='disposition')
```
